# Training data

We will generate negative samples by sampling randomly from the dataset

```python
train_pos = all_stories.copy()
train_pos.reset_index(drop=True, inplace=True)
train_pos['label'] = 1
train_pos.head()
```

```python
train_neg = train_pos.copy()
neg_last_sentence = np.random.choice(train_neg.sentence5, size=len(train_neg))
train_neg['sentence5'] = neg_last_sentence
train_neg.label = 0
train_neg.head()
```

```python
train = pd.concat((train_pos, train_neg))
train = train.sample(frac=1.)  # Shuffle
train.head()
```

```python
train.to_csv('data/train.csv')
```

# Metrics from predictions

```python
# preds = pd.read_csv('eval1/model/predictions.csv')
preds = pd.read_csv('output_nopt/predictions.csv')
preds.head()
```

```python
grouped = preds.groupby('id')

acc = []
acc_ind = []
for storyid, group in grouped:
    true_prob_true = group.loc[group.label == 1].log_prob_true.values[0]
    fake_prob_true = group.loc[group.label == 0].log_prob_true.values[0]
    if true_prob_true > fake_prob_true:
        acc.append(1)
    else:
        acc.append(0)

print('Accuracy:', sum(acc) / len(acc) * 100)
```

```python
preds_probs = preds[['log_prob_fake', 'log_prob_true']]
preds_probs.columns = np.arange(len(preds_probs.columns))
```
